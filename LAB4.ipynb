```python
class Patient:
  def __init__(self, id, name, age, blood_group):
    self.id=id
    self.name=name
    self.age=age
    self.blood_group=blood_group
    self.next=None
    self.prev=None

class WRM:
  def __init__(self):
    self.dummy_head=Patient(None, None, None, None)
    self.dummy_head.next=self.dummy_head
    self.dummy_head.prev=self.dummy_head

  def RegisterPatient(self, id, name, age, bloodgroup):
    new_patient=Patient(id, name, age, bloodgroup)
    last_patient=self.dummy_head.prev
    last_patient.next=new_patient
    new_patient.prev=last_patient
    new_patient.next=self.dummy_head
    self.dummy_head.prev=new_patient

  def ServePatient(self):
    print()
    first_patient=self.dummy_head.next
    if first_patient==self.dummy_head.next.prev:
      print("No patient is served")
      print()
    else:
      first_patient.next.prev=first_patient.prev
      first_patient.prev.next=first_patient.next
      print(f"{first_patient.name} being Served ")
      print()
      return first_patient.name

  def CancelAll(self):
    self.dummy_head.next=self.dummy_head
    self.dummy_head.prev=self.dummy_head
    print()
    print("All appointments have been cancelled.")
    print()

  def CanDoctorGoHome(self):
    return self.dummy_head.next==self.dummy_head

  def ShowAllPatient(self):
    print()
    current_patient=self.dummy_head.next
    if current_patient!=self.dummy_head:
      while current_patient!=self.dummy_head:
        print(current_patient.name,end="  ")
        current_patient=current_patient.next
    else:
      print("No waiting patients")
    print()


  def exitAll(self):
    print()
    print("Waiting Room is closed.")
    print()

wrm = WRM()
while True:
    print(">>>>>>>>>>>CHOOSE AN OPTION<<<<<<<<<<<")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All patients")
    print("4. Can Doctor go Home?")
    print("5. Cancel all Appointment")
    print("6. Close Waiting Room Management System")
    print("=======================================")
    option = int(input("Enter option: "))
    if option == 1:
        id = int(input("Enter id: "))
        name = str(input("Enter name: "))
        age = int(input("Enter age: "))
        bloodgroup = str(input("Enter blood group: "))
        wrm.RegisterPatient(id, name, age, bloodgroup)
        print()
        print("=======================================")
        print("|| SUCCESSFULLY REGISTERED A PATIENT ||")
        print("=======================================")
        print()
    elif option == 2:
        wrm.ServePatient()
    elif option == 3:
        wrm.ShowAllPatient()
        print()
    elif option == 4:
       print()
       if wrm.CanDoctorGoHome():
        print("Yes")
       else:
        print("No")
       print()
    elif option == 5:
        wrm.CancelAll()
    elif option == 6:
        wrm.exitAll()
        break
    else:
      if option >6 or option<=0:
        print()
        print("Invalid Choice..")
        print("Please give choice correctly...")
        print()
```
